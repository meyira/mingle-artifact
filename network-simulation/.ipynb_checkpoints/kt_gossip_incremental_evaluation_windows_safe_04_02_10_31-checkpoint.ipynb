## 0) Setup: repo discovery + CLI flags

```python
import os, sys, re, json, subprocess, time
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

HERE = Path.cwd()
REPO_ROOT = None
for p in [HERE] + list(HERE.parents):
    if (p / "sim").exists():
        REPO_ROOT = p
        break
if REPO_ROOT is None:
    REPO_ROOT = HERE
REPO_ROOT = REPO_ROOT.resolve()
print("Repo root:", REPO_ROOT)

# Prefer repo .venv python if present (no activation needed)
venv_py_win = REPO_ROOT / ".venv" / "Scripts" / "python.exe"
venv_py_unix = REPO_ROOT / ".venv" / "bin" / "python"
if venv_py_win.exists():
    PYTHON_EXE = str(venv_py_win.resolve())
elif venv_py_unix.exists():
    PYTHON_EXE = str(venv_py_unix.resolve())
else:
    PYTHON_EXE = sys.executable
print("Python exe:", PYTHON_EXE)

def run_cmd(cmd, cwd=REPO_ROOT, check=False):
    p = subprocess.run(list(map(str, cmd)), cwd=str(cwd), capture_output=True, text=True)
    if p.stdout.strip():
        print(p.stdout)
    if p.stderr.strip():
        print("STDERR:\n", p.stderr)
    if check and p.returncode != 0:
        raise RuntimeError(f"Command failed rc={p.returncode}")
    return p

def detect_entrypoint():
    p = subprocess.run([PYTHON_EXE, "-m", "sim.run", "--help"], cwd=str(REPO_ROOT), capture_output=True, text=True)
    if p.returncode == 0:
        return ["-m", "sim.run"]
    run_py = REPO_ROOT / "sim" / "run.py"
    if run_py.exists():
        p2 = subprocess.run([PYTHON_EXE, str(run_py), "--help"], cwd=str(REPO_ROOT), capture_output=True, text=True)
        if p2.returncode == 0:
            return [str(run_py)]
    raise RuntimeError("Could not detect simulator entrypoint (sim.run or sim/run.py).")

ENTRY = detect_entrypoint()
print("ENTRY:", ENTRY)

help_p = run_cmd([PYTHON_EXE] + ENTRY + ["--help"], check=True)
HELP_TEXT = (help_p.stdout or "") + "\n" + (help_p.stderr or "")
SUPPORTED_FLAGS = set(re.findall(r"(--[a-zA-Z0-9_]+)", HELP_TEXT))
print("Supported flags:", len(SUPPORTED_FLAGS))

# Path hygiene check
test_path = (REPO_ROOT / "out_path_test" / "a" / "b").resolve()
print("Path test:", test_path)
print("as_posix:", test_path.as_posix())
```

## 2) Runner + derived metrics (Windows-safe)

```python

import re
from pathlib import Path
from itertools import product

def params_to_args(params: dict):
    args = []
    for k, v in params.items():
        flag = f"--{k}"
        if flag not in SUPPORTED_FLAGS:
            continue
        if v is None:
            continue
        if isinstance(v, bool):
            if v:
                args.append(flag)
        else:
            args.extend([flag, str(v)])
    return args

def safe_token(x):
    if isinstance(x, float):
        return f"{x:g}".replace(".", "p")
    s = str(x)
    s = re.sub(r"[^A-Za-z0-9_.-]+", "_", s)
    return s[:24]

def make_outdir_name(cfg, seed):
    parts = [
        safe_token(cfg.get("server_mode","m")),
        safe_token(cfg.get("mode","x")),
        f"n{safe_token(cfg.get('users',0))}",
        f"pg{safe_token(cfg.get('p_gossip',0))}",
        f"a{safe_token(cfg.get('adoption',0))}",
        f"ff{safe_token(cfg.get('fork_frac',0))}",
        f"s{seed}",
    ]
    for k, tag in [("p_drop_crosscut","drop"), ("p_contact_cross","pc"), ("p_group_cross","pgc")]:
        if k in cfg:
            parts.append(f"{tag}{safe_token(cfg.get(k))}")
    return "_".join(parts)

def run_sim(params: dict, outdir):
    outdir = Path(outdir)
    if not outdir.is_absolute():
        outdir = (REPO_ROOT / outdir)
    outdir = outdir.resolve()
    outdir.mkdir(parents=True, exist_ok=True)

    p = dict(params)
    p["outdir"] = str(outdir)

    cmd = [PYTHON_EXE] + ENTRY + params_to_args(p)
    t0 = time.time()
    proc = subprocess.run(list(map(str, cmd)), cwd=str(REPO_ROOT), capture_output=True, text=True)
    wall_s = time.time() - t0
    if proc.stdout.strip():
        print(proc.stdout)
    if proc.stderr.strip():
        print("STDERR:\n", proc.stderr)
    if proc.returncode != 0:
        raise RuntimeError(f"Simulation failed rc={proc.returncode}")
    return outdir, wall_s

def load_run(outdir: Path):
   outdir = Path(outdir)
   summary_path  = outdir / "summary.json"
   timeline_path = outdir / "timeline.csv"
   events_path   = outdir / "events.csv"
   summary  = json.loads(summary_path.read_text(encoding="utf-8")) if summary_path.exists() else {}
   timeline = pd.read_csv(timeline_path) if timeline_path.exists() else pd.DataFrame()
   events   = pd.read_csv(events_path) if events_path.exists() else pd.DataFrame()
   # Normalize timeline columns for backwards/forwards compatibility
   if not timeline.empty:
       # If only old detectors exists, treat it as evidence
       if "evidence_users" not in timeline.columns and "detectors" in timeline.columns:
           timeline["evidence_users"] = timeline["detectors"]
       # If only evidence exists, keep detectors as alias
       if "detectors" not in timeline.columns and "evidence_users" in timeline.columns:
           timeline["detectors"] = timeline["evidence_users"]
       # Awareness column may be missing on older runs
       if "aware_users" not in timeline.columns:
           timeline["aware_users"] = np.nan
   return summary, timeline, events

def time_to_fraction(timeline: pd.DataFrame, users: int, dt_min: float, frac: float, col: str):
   if timeline.empty or users <= 0:
       return np.nan
   if col not in timeline.columns or "tick" not in timeline.columns:
       return np.nan
   target = frac * users
   arr = timeline[col].to_numpy()
   ticks = timeline["tick"].to_numpy()
   idx = np.where(arr >= target)[0]
   if len(idx) == 0:
       return np.nan
   return float(ticks[idx[0]]) * dt_min

def derived_metrics(summary: dict, timeline: pd.DataFrame, fallback: dict, detection_policy="strict"):
   users = int(summary.get("users", fallback.get("users", 0)))
   dt_min = float(summary.get("dt_min", fallback.get("dt_min", 5)))
   fork_hour = float(summary.get("fork_hour", fallback.get("fork_hour", 0)))
   fork_offset_min = fork_hour * 60.0
   # Proof outcomes
   pr_ok   = int(summary.get("proof_ok", 0))
   pr_inc  = int(summary.get("proof_inconsistent", 0))
   pr_with = int(summary.get("proof_withhold", 0))
   pr_to   = int(summary.get("proof_timeout", 0))
   evidence_events = pr_ok + pr_inc + pr_with + pr_to
   # Strict vs soft detection
   if detection_policy == "strict":
       detected = (pr_inc > 0)
   else:
       detected = ((pr_inc + pr_with + pr_to) > 0)
   # New summary fields (from patched sim). Fall back if absent.
   first_aw = summary.get("first_awareness_tick", None)
   first_ev = summary.get("first_evidence_tick", summary.get("first_detection_tick", None))
   def tick_to_latency(first_tick):
       if first_tick in (None, ""):
           return np.nan
       try:
           return max(0.0, float(first_tick) * dt_min - fork_offset_min)
       except Exception:
           return np.nan
   lat_aw_min = tick_to_latency(first_aw)
   lat_ev_min = tick_to_latency(first_ev)
   # End coverage using summary if present; else use timeline if present
   aware_end = summary.get("aware_end", None)
   evidence_end = summary.get("evidence_end", None)
   if aware_end is None and not timeline.empty and "aware_users" in timeline.columns and users > 0:
       aware_end = float(timeline["aware_users"].iloc[-1])
   if evidence_end is None and not timeline.empty and "evidence_users" in timeline.columns and users > 0:
       evidence_end = float(timeline["evidence_users"].iloc[-1])
   aware_cov_end = (float(aware_end) / users) if (aware_end is not None and users > 0) else np.nan
   evid_cov_end  = (float(evidence_end) / users) if (evidence_end is not None and users > 0) else np.nan
   # Time-to-X% AFTER fork, for both layers
   t10_aw_abs = time_to_fraction(timeline, users, dt_min, 0.10, col="aware_users")
   t50_aw_abs = time_to_fraction(timeline, users, dt_min, 0.50, col="aware_users")
   t10_ev_abs = time_to_fraction(timeline, users, dt_min, 0.10, col="evidence_users")
   t50_ev_abs = time_to_fraction(timeline, users, dt_min, 0.50, col="evidence_users")
   def to_after_fork(t_abs):
       if not np.isfinite(t_abs):
           return np.nan
       return max(0.0, float(t_abs) - fork_offset_min)
   t10_aw = to_after_fork(t10_aw_abs)
   t50_aw = to_after_fork(t50_aw_abs)
   t10_ev = to_after_fork(t10_ev_abs)
   t50_ev = to_after_fork(t50_ev_abs)
   ## Overhead (sim only provides total_messages; use it as denominator)
   total_gossip_bytes = float(summary.get("total_gossip_bytes", 0) or 0)
   total_msgs  = float(summary.get("total_messages", 0) or 0)
   # If sends/deliveries are not tracked, fall back to total_messages
   total_sends = float(summary.get("total_sends", total_msgs) or total_msgs)
   total_deliv = float(summary.get("total_deliveries", total_msgs) or total_msgs)
   bps = total_gossip_bytes / max(1.0, total_sends)
   bpd = total_gossip_bytes / max(1.0, total_deliv)
   # Crosscut counters
   x_deliv = float(summary.get("crosscut_deliveries", 0) or 0)
   x_drop  = float(summary.get("crosscut_dropped", 0) or 0)
   crosscuts_eff = max(0.0, x_deliv - x_drop)
   crosscut_rate_eff = crosscuts_eff / max(1.0, total_deliv)
   out = dict(
       server_mode=summary.get("server_mode", fallback.get("server_mode")),
       mode=summary.get("mode", fallback.get("mode")),
       users=users,
       hours=float(summary.get("hours", fallback.get("hours", np.nan))),
       dt_min=dt_min,
       p_gossip=float(summary.get("p_gossip", fallback.get("p_gossip", np.nan))),
       adoption=float(summary.get("adoption", fallback.get("adoption", np.nan))),
       fork_frac=float(summary.get("fork_frac", fallback.get("fork_frac", np.nan))),
       fork_hour=float(summary.get("fork_hour", fallback.get("fork_hour", np.nan))),
       detected=bool(detected),
       # Awareness vs evidence
       lat_aw_min=lat_aw_min,
       lat_ev_min=lat_ev_min,
       aware_cov_end=aware_cov_end,
       evid_cov_end=evid_cov_end,
       t10_aw_med=t10_aw,
       t50_aw_med=t50_aw,
       t10_ev_med=t10_ev,
       t50_ev_med=t50_ev,
       # Overhead
       bytes_per_send=bps,
       bytes_per_delivery=bpd,
       # Crosscuts + evidence events
       crosscuts=crosscuts_eff,
       #crosscut_dropped=crosscut_dropped,
       crosscut_rate=crosscut_rate_eff,
       proof_inconsistent=pr_inc,
       evidence_events=evidence_events,
       seed=int(summary.get("seed", fallback.get("seed", -1))),
   )
   # Keep optional knobs if present
   for k in ["p_contact_cross","p_group_cross","p_drop_crosscut","sticky_victims",
             "msg_rate_1to1","msg_rate_group","groups","avg_group_size",
             "proof_withhold_rate","proof_timeout_rate","invalid_proof_rate"]:
       if k in summary or k in fallback:
           out[k] = summary.get(k, fallback.get(k))
   return out
 
def run_grid(base: dict, sweep: dict, replicates: int, seed_base: int, outroot: Path, detection_policy="strict"):
    outroot = Path(outroot)
    if not outroot.is_absolute():
        outroot = (REPO_ROOT / outroot)
    outroot = outroot.resolve()
    outroot.mkdir(parents=True, exist_ok=True)

    rows = []
    keys = list(sweep.keys())

    for vals in product(*[sweep[k] for k in keys]):
        cfg = dict(base)
        cfg.update({k:v for k,v in zip(keys, vals)})
        for r in range(replicates):
            seed = seed_base + r
            cfg_run = dict(cfg); cfg_run["seed"] = seed
            outdir = outroot / make_outdir_name(cfg_run, seed)

            outdir_path, wall_s = run_sim(cfg_run, outdir)
            summary, timeline, _ = load_run(outdir_path)
            dm = derived_metrics(summary, timeline, fallback=cfg_run, detection_policy=detection_policy)
            dm["wall_s"] = wall_s
            dm["run_dir"] = outdir_path.as_posix()
            rows.append(dm)
    return pd.DataFrame(rows)

def aggregate(df_raw: pd.DataFrame, group_cols):
   return df_raw.groupby(group_cols, dropna=False).agg(
       reps=("run_dir","count"),
       wall_s_med=("wall_s","median"),
       # End coverage
       aware_cov_end_mean=("aware_cov_end","mean"),
       evid_cov_end_mean=("evid_cov_end","mean"),
       # Detection rate (strict/soft depending on policy used)
       detected_rate=("detected","mean"),
       # Latencies
       lat_aw_med=("lat_aw_min","median"),
       lat_ev_med=("lat_ev_min","median"),
       # Time-to-X% after fork
       t10_aw_med=("t10_aw_med","median"),
       t50_aw_med=("t50_aw_med","median"),
       t10_ev_med=("t10_ev_med","median"),
       t50_ev_med=("t50_ev_med","median"),
       # Overhead
       bpd_mean=("bytes_per_delivery","mean"),
       bps_mean=("bytes_per_send","mean"),
       # Crosscuts / evidence
       crosscuts_mean=("crosscuts","mean"),
       evidence_mean=("evidence_events","mean"),
   ).reset_index()
```

```python
# =========================
# Topology & workload validation (Table X helper) - SIGNATURE-SAFE FULL CELL
# Put this cell right after initialization/imports, before experiments/sweeps.
# =========================
import numpy as np
import pandas as pd
from pathlib import Path
import importlib
import inspect
SURVEY_CSV = REPO_ROOT / "results-survey899426.csv"
if not SURVEY_CSV.exists():
   SURVEY_CSV = Path.cwd() / "results-survey899426.csv"
print("Survey CSV:", SURVEY_CSV)
def qstats(x):
   x = pd.to_numeric(pd.Series(x).dropna(), errors="coerce").dropna()
   if len(x) == 0:
       return dict(n=0, mean=np.nan, p50=np.nan, p75=np.nan, p90=np.nan, p95=np.nan, max=np.nan)
   return dict(
       n=int(len(x)),
       mean=float(x.mean()),
       p50=float(x.quantile(0.50)),
       p75=float(x.quantile(0.75)),
       p90=float(x.quantile(0.90)),
       p95=float(x.quantile(0.95)),
       max=float(x.max()),
   )
def make_labels(n, frac, rng):
   frac = float(frac)
   if frac <= 0:
       return None
   return (rng.random(n) < frac).astype(int)
def iter_group_members(groups_obj):
   if groups_obj is None:
       return
   if isinstance(groups_obj, dict):
       for _, members in groups_obj.items():
           yield members
       return
   if isinstance(groups_obj, (list, tuple)):
       for item in groups_obj:
           if isinstance(item, tuple) and len(item) == 2 and isinstance(item[1], (list, tuple, np.ndarray, set)):
               yield item[1]
           else:
               yield item
       return
   try:
       for item in groups_obj:
           yield item
   except TypeError:
       return
def flatten_members(members):
   if members is None:
       return []
   if isinstance(members, set):
       members = list(members)
   elif isinstance(members, np.ndarray):
       members = members.tolist()
   # one-level flatten if nested
   if isinstance(members, (list, tuple)) and len(members) > 0 and isinstance(members[0], (list, tuple, np.ndarray, set)):
       flat = []
       for x in members:
           if isinstance(x, set): x = list(x)
           if isinstance(x, np.ndarray): x = x.tolist()
           try:
               flat.extend(list(x))
           except TypeError:
               flat.append(x)
       members = flat
   out = []
   if isinstance(members, (list, tuple)):
       for x in members:
           try:
               out.append(int(x))
           except Exception:
               pass
   else:
       try:
           out.append(int(members))
       except Exception:
           pass
   return out
def call_with_signature(fn, desired: dict):
   """
   Call fn by mapping desired keys to whatever parameter names fn accepts.
   This avoids passing 'n' to the wrong parameter and creating ID/label mismatches.
   """
   sig = inspect.signature(fn)
   params = sig.parameters
   # candidate name sets
   name_map = {
       "n": ["n", "N", "users", "n_users", "n_clients", "num_users", "num_clients"],
       "avg_deg": ["avg_deg", "avg_degree", "avg_contacts", "mean_deg", "mean_degree"],
       "groups": ["groups", "n_groups", "num_groups"],
       "avg_size": ["avg_size", "avg_group_size", "mean_size", "group_size_mean"],
       "rng": ["rng", "random", "np_rng", "generator"],
       "labels": ["labels", "cohorts", "victims", "victim_labels"],
       "p_cross": ["p_cross", "p_contact_cross", "p_group_cross", "cross_p", "p_mix"],
   }
   kwargs = {}
   for key, val in desired.items():
       if key not in name_map:
           continue
       for cand in name_map[key]:
           if cand in params:
               kwargs[cand] = val
               break
   # Some fns may require positional-only params; try kwargs first then fall back.
   try:
       return fn(**kwargs)
   except TypeError:
       # last resort: try passing the most common positional ordering
       # contacts: (n, avg_deg, rng, labels?, p_cross?)
       # groups:   (n, groups, avg_size, rng, labels?, p_cross?)
       pos = []
       for k in ["n", "avg_deg", "groups", "avg_size", "rng"]:
           if k in desired:
               pos.append(desired[k])
       return fn(*pos)
def summarize_topology(cfg, seed=123, debug=True):
   g = importlib.import_module("sim.graph")
   gen_contacts = getattr(g, "gen_contacts", None) or getattr(g, "make_contacts", None) or getattr(g, "build_contacts", None)
   gen_groups   = getattr(g, "gen_groups", None)   or getattr(g, "make_groups", None)   or getattr(g, "build_groups", None)
   if gen_contacts is None or gen_groups is None:
       raise ImportError(
           "Could not find gen_contacts/gen_groups in sim.graph. "
           f"Available symbols: {[k for k in dir(g) if 'contact' in k.lower() or 'group' in k.lower()]}"
       )
   n_cfg = int(cfg["users"])
   avg_deg = float(cfg["avg_contacts"])
   n_groups = int(cfg["groups"])
   avg_size = float(cfg["avg_group_size"])
   pcc = float(cfg.get("p_contact_cross", 1.0))
   pgc = float(cfg.get("p_group_cross", 1.0))
   fork_frac = float(cfg.get("fork_frac", 0.0))
   rng = np.random.default_rng(seed)
   labels = make_labels(n_cfg, fork_frac, rng)
   contacts = call_with_signature(gen_contacts, dict(
       n=n_cfg, avg_deg=avg_deg, rng=rng, labels=labels, p_cross=pcc
   ))
   groups_obj = call_with_signature(gen_groups, dict(
       n=n_cfg, groups=n_groups, avg_size=avg_size, rng=rng, labels=labels, p_cross=pgc
   ))
   # --- infer effective N from realized IDs ---
   # contacts is expected to be list-like indexed by user id
   n_contacts = len(contacts) if hasattr(contacts, "__len__") else n_cfg
   max_id = n_contacts - 1
   # scan contacts
   try:
       for u in range(n_contacts):
           for v in contacts[u]:
               try:
                   max_id = max(max_id, int(v))
               except Exception:
                   pass
   except Exception:
       pass
   # scan groups
   for gmem in iter_group_members(groups_obj):
       mem = flatten_members(gmem)
       if mem:
           max_id = max(max_id, max(mem))
   n_eff = max(n_cfg, n_contacts, max_id + 1)
   # if labels were created for the wrong N, regenerate at correct size
   if labels is not None and len(labels) != n_eff:
       rng2 = np.random.default_rng(seed + 999)
       labels = make_labels(n_eff, fork_frac, rng2)
   if debug:
       print("n_cfg:", n_cfg, "n_contacts:", n_contacts, "max_id:", max_id, "n_eff:", n_eff)
       print("labels len:", None if labels is None else len(labels))
       print("groups_obj type:", type(groups_obj))
   # degrees (only for users present in contacts list)
   degs = np.array([len(contacts[i]) for i in range(n_contacts)], dtype=int)
   # group sizes
   g_sizes = []
   for gmem in iter_group_members(groups_obj):
       mem = flatten_members(gmem)
       # keep only members within label range
       mem = [m for m in mem if 0 <= m < n_eff]
       if mem:
           g_sizes.append(len(mem))
   gsz = np.array(g_sizes, dtype=int) if g_sizes else np.array([], dtype=int)
   # observed crosscut fraction among contact choices
   cross_contact_frac = np.nan
   if labels is not None:
       tot = 0
       cross = 0
       for u in range(n_contacts):
           lu = labels[u]
           nbrs = contacts[u]
           for v in nbrs:
               try:
                   v = int(v)
               except Exception:
                   continue
               if 0 <= v < n_eff:
                   tot += 1
                   if labels[v] != lu:
                       cross += 1
       cross_contact_frac = (cross / tot) if tot > 0 else np.nan
   # observed fraction of crosscut members inside groups
   cross_group_frac = np.nan
   if labels is not None and len(g_sizes):
       tot_m = 0
       cross_m = 0
       for gmem in iter_group_members(groups_obj):
           mem = flatten_members(gmem)
           mem = [m for m in mem if 0 <= m < n_eff]
           if not mem:
               continue
           lbls = labels[np.array(mem, dtype=int)]
           dom = 1 if lbls.sum() >= (len(lbls) / 2) else 0
           tot_m += len(mem)
           cross_m += int((lbls != dom).sum())
       cross_group_frac = (cross_m / tot_m) if tot_m > 0 else np.nan
   topo = {
       "N_cfg": n_cfg,
       "N_eff": n_eff,
       "avg_contacts_param": avg_deg,
       "groups_param": n_groups,
       "avg_group_size_param": avg_size,
       "p_contact_cross_param": pcc,
       "p_group_cross_param": pgc,
       "fork_frac_labels": fork_frac if labels is not None else 0.0,
       "deg_mean": float(degs.mean()) if len(degs) else np.nan,
       "deg_p50": float(np.quantile(degs, 0.50)) if len(degs) else np.nan,
       "deg_p95": float(np.quantile(degs, 0.95)) if len(degs) else np.nan,
       "deg_max": float(degs.max()) if len(degs) else np.nan,
       "gsize_mean": float(gsz.mean()) if len(gsz) else np.nan,
       "gsize_p50": float(np.quantile(gsz, 0.50)) if len(gsz) else np.nan,
       "gsize_p95": float(np.quantile(gsz, 0.95)) if len(gsz) else np.nan,
       "gsize_max": float(gsz.max()) if len(gsz) else np.nan,
       "cross_contact_frac_obs": float(cross_contact_frac) if np.isfinite(cross_contact_frac) else np.nan,
       "cross_group_frac_obs": float(cross_group_frac) if np.isfinite(cross_group_frac) else np.nan,
   }
   return topo
# --- Survey anchors (best-effort column matching) ---
survey_stats = {}
if SURVEY_CSV.exists():
   df_s = pd.read_csv(SURVEY_CSV)
   def pick_col(patterns):
       for p in patterns:
           hits = [c for c in df_s.columns if p.lower() in c.lower()]
           if hits:
               return hits[0]
       return None
   col_week  = pick_col(["unique contacts per week", "weekly unique", "last week"])
   col_day   = pick_col(["unique contacts per day", "yesterday", "daily unique"])
   col_phone = pick_col(["phone contacts", "contacts in phone"])
   group_cols = [c for c in df_s.columns if ("participants" in c.lower()) or ("group size" in c.lower())]
   if col_week:  survey_stats["survey_unique_peers_week"] = qstats(df_s[col_week])
   if col_day:   survey_stats["survey_unique_peers_day"] = qstats(df_s[col_day])
   if col_phone: survey_stats["survey_phone_contacts"] = qstats(df_s[col_phone])
   if group_cols:
       all_g = pd.concat([pd.to_numeric(df_s[c], errors="coerce") for c in group_cols], ignore_index=True)
       survey_stats["survey_group_sizes_all"] = qstats(all_g)
print("Survey anchors found:", list(survey_stats.keys()))
for k, v in survey_stats.items():
   print(k, ":", v)
# --- Choose baseline config to validate ---
CFG_TO_VALIDATE = dict(BASE_SMALL)      # or BASE_HARD / BASE_HIMIX
CFG_TO_VALIDATE.setdefault("fork_frac", 0.02)         # enable labels for crosscut measurement
CFG_TO_VALIDATE.setdefault("p_contact_cross", 0.005)
CFG_TO_VALIDATE.setdefault("p_group_cross", 0.005)
topo = summarize_topology(CFG_TO_VALIDATE, seed=4242, debug=True)
topo_df = pd.DataFrame([topo])
print("\n=== Topology validation (measured from generator) ===")
display(topo_df)
OUT_VAL = OUT_ROOT / "validation"
OUT_VAL.mkdir(parents=True, exist_ok=True)
topo_df.to_csv(OUT_VAL / "topology_validation.csv", index=False)
print("Wrote:", OUT_VAL / "topology_validation.csv")
```

```python

def one_run(name, cfg):
   outdir = OUT_ROOT / "sanity" / name
   outdir_path, wall_s = run_sim(cfg, outdir)
   summary, timeline, _ = load_run(outdir_path)
   dm = derived_metrics(summary, timeline, fallback=cfg, detection_policy="strict")
    # ---- Normalize metric names for tables/plots ----
    # Use evidence metrics as the detection metrics in the baseline.
   dm["coverage_end"] = dm.get("evid_cov_end", dm.get("coverage_end", np.nan))
    # Depending on notebook version, t10/t50 may be stored as *_ev_med (minutes) or *_ev_min
   dm["t10_min"] = dm.get("t10_ev_med", dm.get("t10_ev_min", dm.get("t10_min", np.nan)))
   dm["t50_min"] = dm.get("t50_ev_med", dm.get("t50_ev_min", dm.get("t50_min", np.nan)))
   # Overhead alias (some versions call it bytes_per_delivery, some bpd)
   if "bytes_per_delivery" not in dm:
      dm["bytes_per_delivery"] = dm.get("bpd", dm.get("bpd_mean", np.nan))
        
   dm["name"] = name
   dm["wall_s"] = wall_s
   # Pull these directly from summary.json (they exist there)
   dm["proof_inconsistent"] = int(summary.get("proof_inconsistent", 0))
   dm["proof_ok"] = int(summary.get("proof_ok", 0))
   dm["proof_withhold"] = int(summary.get("proof_withhold", 0))
   dm["proof_timeout"] = int(summary.get("proof_timeout", 0))
   # Crosscut diagnostics
   total_deliv = float(summary.get("total_deliveries", summary.get("total_messages", 0)) or 0)
   crosscuts = float(summary.get("crosscut_deliveries", 0))
   x_deliv = float(summary.get("crosscut_deliveries", 0))
   x_drop  = float(summary.get("crosscut_dropped", 0))
   # Treat "delivered crosscuts" as delivered-minus-dropped (robust even if simulator double-counts)
   x_effective = max(0.0, x_deliv - x_drop)
   dm["crosscuts_deliv"] = x_deliv
   dm["crosscuts_drop"] = x_drop
   dm["crosscuts_eff"] = x_effective
   dm["crosscut_rate_eff"] = (x_effective / total_deliv) if total_deliv > 0 else np.nan
    
   dm["proof_per_crosscut"] = (dm["proof_inconsistent"] / crosscuts) if crosscuts > 0 else np.nan
   return dm
cases = []
cases.append(one_run(
   "honest",
   dict(BASE_SMALL, server_mode="honest", p_gossip=0.2, adoption=0.5)
))
cases.append(one_run(
   "fork_no_gossip",
   dict(BASE_SMALL, server_mode="permanent_fork", p_gossip=0.0, adoption=0.0)
))
cases.append(one_run(
   "fork_with_gossip",
   dict(BASE_SMALL, server_mode="permanent_fork", p_gossip=0.2, adoption=0.5)
))
cases.append(one_run(
   "fork_with_gossip_highmix",
   dict(
       BASE_SMALL,
       hours=6.0, fork_hour=3.0, fork_duration=2.0,
       server_mode="permanent_fork",
       p_gossip=0.2, adoption=1.0,
       msg_rate_1to1=0.20,
       msg_rate_group=0.08,
       p_contact_cross=0.05,
       p_group_cross=0.05
   )
))
cases.append(one_run(
   "fork_with_gossip_censored",
   dict(
       BASE_SMALL,
       hours=6.0, fork_hour=3.0, fork_duration=2.0,
       server_mode="permanent_fork",
       p_gossip=0.2, adoption=1.0,
       msg_rate_1to1=0.20,
       msg_rate_group=0.08,
       p_contact_cross=0.05,
       p_group_cross=0.05,
       p_drop_crosscut=1.0
   )
))
df_sanity2 = pd.DataFrame(cases)
df_sanity2.to_csv(OUT_ROOT / "sanity_ladder.csv", index=False)
# Show key columns (now guaranteed to exist)
cols = [
   "name","server_mode","p_gossip","adoption",
   "detected","coverage_end","t10_min","t50_min",
   "crosscuts_eff","crosscut_rate_eff","proof_inconsistent","proof_per_crosscut",
   "bytes_per_delivery","wall_s"
]
df_sanity2[cols]
```

```python
# =========================
# Aggregate + NaN-safe IQR (for latency)
# =========================
import numpy as np
def aggregate_iqr_safe(df_raw: pd.DataFrame, group_cols):
   def _finite(x):
       x = np.asarray(x, dtype=float)
       return x[np.isfinite(x)]
   def q25(x):
       x = _finite(x);  return float(np.quantile(x, 0.25)) if x.size else np.nan
   def q75(x):
       x = _finite(x);  return float(np.quantile(x, 0.75)) if x.size else np.nan
   def med(x):
       x = _finite(x);  return float(np.median(x)) if x.size else np.nan
   g = df_raw.groupby(group_cols, dropna=False)
   out = g.agg(
       reps=("run_dir","count"),
       detected_rate=("detected","mean"),
       lat_ev_med=("lat_ev_min", med),
       lat_ev_q25=("lat_ev_min", q25),
       lat_ev_q75=("lat_ev_min", q75),
       bpd_mean=("bytes_per_delivery","mean"),
       bps_mean=("bytes_per_send","mean"),
       evidence_mean=("evidence_events","mean"),
       crosscuts_mean=("crosscuts","mean"),
   ).reset_index()
   return out
GROUP_COLS_SHOW = [
   "users","adoption","p_gossip",
   "server_mode","mode",
   "fork_frac","p_contact_cross","p_group_cross",
   "p_drop_crosscut"
]
df_agg_show = aggregate_iqr_safe(df_raw_show, GROUP_COLS_SHOW)\
               .sort_values(["users","adoption","p_gossip"])
df_agg_show.to_csv(OUT_SHOW / "agg.csv", index=False)
display(df_agg_show.head(15))
print("Wrote:", (OUT_SHOW/"agg.csv").resolve())
```

```python
# =========================
# Paper-ready plots for Typical regime (monotone det, Pareto frontier, min p_gossip curve)
# =========================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
PLOTS_SHOWCASE = Path(OUT_TYPICAL) / "plots_showcase"
PLOTS_SHOWCASE.mkdir(parents=True, exist_ok=True)
COL_USERS, COL_ADOPT, COL_PG = "users", "adoption", "p_gossip"
COL_DET, COL_LAT, COL_OVH = "detected_rate", "lat_ev_med", "bpd_mean"
def pareto_frontier(df, x, y):
   d = df[[x, y]].dropna().sort_values(x).copy()
   best = np.inf
   keep = []
   for _, r in d.iterrows():
       if float(r[y]) < best:
           best = float(r[y])
           keep.append(True)
       else:
           keep.append(False)
   return d.loc[keep].copy()
def isotonic_fit_monotone(x, y):
# PAVA for non-decreasing y
   x = np.asarray(x, dtype=float)
   y = np.asarray(y, dtype=float)
   m = np.isfinite(x) & np.isfinite(y)
   x, y = x[m], y[m]
   if x.size == 0:
       return np.array([]), np.array([])
   order = np.argsort(x)
   x, y = x[order], y[order]
   levels = y.copy()
   w = np.ones_like(levels)
   i = 0
   while i < len(levels) - 1:
       if levels[i] <= levels[i+1]:
           i += 1
           continue
       new_level = (levels[i]*w[i] + levels[i+1]*w[i+1]) / (w[i] + w[i+1])
       levels[i] = new_level
       w[i] += w[i+1]
       levels = np.delete(levels, i+1)
       w = np.delete(w, i+1)
       x = np.delete(x, i+1)
       i = max(i-1, 0)
   return x, levels
def min_pg_for_threshold(x_step, y_step, thr=0.95):
   if len(x_step) == 0:
       return np.nan
   for xi, yi in zip(x_step, y_step):
       if yi >= thr:
           return float(xi)
   return np.nan
def plot_showcase_for_users(df_all, users, det_thr=0.95, adoption_order=None):
   dU = df_all[df_all[COL_USERS] == users].copy()
   dU[COL_ADOPT] = dU[COL_ADOPT].astype(float)
   dU[COL_PG] = dU[COL_PG].astype(float)
   if adoption_order is None:
       adoption_order = sorted(dU[COL_ADOPT].dropna().unique())
   fig = plt.figure(figsize=(3.45, 6.7))
   gs = fig.add_gridspec(3, 1, height_ratios=[1.1, 1.25, 0.85], hspace=0.35)
   ax1 = fig.add_subplot(gs[0, 0])
   ax2 = fig.add_subplot(gs[1, 0])
   ax3 = fig.add_subplot(gs[2, 0])
# Panel 1: detection curves
   min_pg = []
   for a in adoption_order:
       s = dU[dU[COL_ADOPT] == a].sort_values(COL_PG)
       x = s[COL_PG].to_numpy()
       y = s[COL_DET].to_numpy()
       xs, ys = isotonic_fit_monotone(x, y)
# raw points (small)
       m = np.isfinite(x) & np.isfinite(y)
       ax1.plot(x[m], y[m], marker="o", linestyle="None", markersize=2.8, alpha=0.55)
# monotone step
       if len(xs):
           ax1.step(xs, ys, where="post", linewidth=1.4, label=f"adopt={a:g}")
       min_pg.append((a, min_pg_for_threshold(xs, ys, det_thr)))
   ax1.set_xscale("log")
   ax1.axhline(det_thr, linestyle="--", linewidth=1.0)
   ax1.set_ylim(-0.02, 1.02)
   ax1.set_ylabel("Detection prob.")
   ax1.set_xlabel(r"$p_{\mathrm{gossip}}$ (log)")
   ax1.set_title(f"Typical regime, N={users}")
   ax1.grid(True, alpha=0.25, which="both")
   ax1.legend(fontsize=7, frameon=True, ncol=2, loc="lower right")
# Panel 2: latency vs overhead for det>=thr + Pareto
   rel = dU[dU[COL_DET] >= det_thr].copy()
   for a in adoption_order:
       s = rel[rel[COL_ADOPT] == a].copy()
       x = s[COL_OVH].astype(float).to_numpy()
       y = s[COL_LAT].astype(float).to_numpy()
       m = np.isfinite(x) & np.isfinite(y)
       if m.any():
           ax2.scatter(x[m], y[m], s=16, alpha=0.75, label=f"adopt={a:g}")
   front = pareto_frontier(rel, COL_OVH, COL_LAT)
   if not front.empty:
       ax2.plot(front[COL_OVH], front[COL_LAT], linewidth=1.6)
   ax2.set_xlabel("Overhead (B / delivered msg)")
   ax2.set_ylabel("Evidence latency (min)")
   ax2.grid(True, alpha=0.25)
   ax2.legend(fontsize=7, frameon=True, ncol=2, loc="upper right")
# Panel 3: min p_gossip for det>=thr
   mp = pd.DataFrame(min_pg, columns=[COL_ADOPT, "min_p_gossip"])
   mp = mp[np.isfinite(mp["min_p_gossip"])].sort_values(COL_ADOPT)
   if not mp.empty:
       ax3.plot(mp[COL_ADOPT], mp["min_p_gossip"], marker="o", linewidth=1.5)
   ax3.set_yscale("log")
   ax3.set_xlabel("Adoption")
   ax3.set_ylabel(r"min $p_{\mathrm{gossip}}$")
   ax3.grid(True, alpha=0.25, which="both")
   fig.tight_layout()
   outbase = PLOTS_SHOWCASE / f"fig_typical_showcase_N{users}"
   fig.savefig(outbase.with_suffix(".pdf"))
   fig.savefig(outbase.with_suffix(".png"), dpi=300)
   plt.show()
   print("Wrote:", outbase.with_suffix(".pdf").resolve())
for U in sorted(df_agg_typical[COL_USERS].unique()):
   plot_showcase_for_users(df_agg_typical, users=int(U), det_thr=0.95)
```

```python
# =========================
# OPTIONAL: Two-stage adaptive sweep for ultra-smooth thresholds (reruns)
# Stage 1 coarse -> estimate threshold band
# Stage 2 dense p_gossip grid around the band
# =========================
import numpy as np
import pandas as pd
from pathlib import Path
OUT_ADAPT = OUT_ROOT / "paper_sweeps_typical_adaptive"
OUT_ADAPT.mkdir(parents=True, exist_ok=True)
BASE_TYPICAL_ADAPT = dict(BASE_TYPICAL)  # reuse from Cell A
USERS_GRID = [12000]                     # focus for paper headline
ADOPTION_GRID = [0.05, 0.10, 0.20, 0.30, 0.60]
PG_COARSE = [0.002, 0.005, 0.01, 0.02, 0.05, 0.10, 0.20]
REPS1 = 12
REPS2 = 20
# ---- Stage 1 ----
df_raw_1 = run_grid(
   BASE_TYPICAL_ADAPT,
   dict(users=USERS_GRID, adoption=ADOPTION_GRID, p_gossip=PG_COARSE),
   replicates=REPS1, seed_base=30000,
   outroot=OUT_ADAPT / "stage1",
   detection_policy="strict",
)
# quick aggregate for threshold estimate
g1 = df_raw_1.groupby(["users","adoption","p_gossip"], dropna=False)["detected"].mean().reset_index()
g1 = g1.sort_values(["adoption","p_gossip"])
# estimate a per-adoption p_gossip band around det~0.95
TARGET = 0.95
bands = []
for a in sorted(g1["adoption"].unique()):
   s = g1[g1["adoption"] == a].copy()
# first pg where det >= TARGET
   ok = s[s["detected"] >= TARGET]
   if ok.empty:
       continue
   pg_star = float(ok["p_gossip"].iloc[0])
   bands.append((a, pg_star))
bands_df = pd.DataFrame(bands, columns=["adoption","pg_star"])
print("Estimated pg_star per adoption:\n", bands_df)
# ---- Stage 2: dense grid around pg_star (x/2 to x*2 clipped to [0.001,0.2]) ----
dense_jobs = []
for _, r in bands_df.iterrows():
   a = float(r["adoption"])
   pg = float(r["pg_star"])
   lo = max(0.001, pg/2.0)
   hi = min(0.2, pg*2.0)
   pg_dense = np.unique(np.round(np.geomspace(lo, hi, 14), 6)).tolist()
   dense_jobs.append((a, pg_dense))
# run stage2 per adoption to keep folders manageable
df_raw_2_all = []
for a, pg_dense in dense_jobs:
   out = OUT_ADAPT / "stage2" / f"adopt_{a:g}"
   out.mkdir(parents=True, exist_ok=True)
   df_raw_2 = run_grid(
       BASE_TYPICAL_ADAPT,
       dict(users=USERS_GRID, adoption=[a], p_gossip=pg_dense),
       replicates=REPS2, seed_base=40000 + int(a*1000),
       outroot=out,
       detection_policy="strict",
   )
   df_raw_2_all.append(df_raw_2)
df_raw_2 = pd.concat(df_raw_2_all, ignore_index=True) if df_raw_2_all else pd.DataFrame()
df_raw_adapt = pd.concat([df_raw_1, df_raw_2], ignore_index=True)
df_raw_adapt.to_csv(OUT_ADAPT / "raw_adaptive.csv", index=False)
print("Wrote:", (OUT_ADAPT / "raw_adaptive.csv").resolve(), "Rows:", len(df_raw_adapt))
df_raw_adapt.head()
```

```python
# =========================
# Typical regime plots (paper-ready): detection, latency, and latency-vs-overhead
# =========================
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
PLOTS_TYP = Path(OUT_TYP) / "plots"
PLOTS_TYP.mkdir(parents=True, exist_ok=True)
def _savefig(basepath: Path):
   plt.tight_layout()
   plt.savefig(basepath.with_suffix(".pdf"))
   plt.savefig(basepath.with_suffix(".png"), dpi=240)
   print("Wrote:", basepath.with_suffix(".pdf").resolve())
def plot_det_vs_pg(df, mix_level, outpath, users=12000):
   sub = df[(df["users"] == users) & np.isclose(df["mix_level"], mix_level)].copy()
   sub = sub.sort_values(["adoption","p_gossip"])
   plt.figure(figsize=(3.35, 2.35))
   for a in sorted(sub["adoption"].unique()):
       s = sub[sub["adoption"] == a].sort_values("p_gossip")
       x = s["p_gossip"].astype(float).to_numpy()
       y = s["detected_rate"].astype(float).to_numpy()
       m = np.isfinite(x) & np.isfinite(y)
       plt.plot(x[m], y[m], marker="o", linewidth=1.3, markersize=3.5, label=f"adopt={a:g}")
   plt.xscale("log")
   plt.ylim(-0.02, 1.02)
   plt.xlabel(r"$p_{\mathrm{gossip}}$")
   plt.ylabel("Detection prob.")
   plt.title(fr"Typical regime: detection (mix={mix_level:g})")
   plt.grid(True, alpha=0.25, which="both")
   plt.legend(fontsize=6.5, frameon=True, ncol=3, loc="lower right")
   _savefig(outpath)
   plt.show()
def plot_lat_vs_pg(df, mix_level, outpath, users=12000):
   sub = df[(df["users"] == users) & np.isclose(df["mix_level"], mix_level)].copy()
   sub = sub.sort_values(["adoption","p_gossip"])
   plt.figure(figsize=(3.35, 2.35))
   for a in sorted(sub["adoption"].unique()):
       s = sub[sub["adoption"] == a].sort_values("p_gossip")
       x  = s["p_gossip"].astype(float).to_numpy()
       y  = s["lat_ev_med"].astype(float).to_numpy()
       lo = s["lat_ev_q25"].astype(float).to_numpy()
       hi = s["lat_ev_q75"].astype(float).to_numpy()
       m = np.isfinite(x) & np.isfinite(y)
       plt.plot(x[m], y[m], marker="o", linewidth=1.3, markersize=3.5, label=f"adopt={a:g}")
       mb = np.isfinite(x) & np.isfinite(lo) & np.isfinite(hi)
       if mb.any():
           plt.fill_between(x[mb], lo[mb], hi[mb], alpha=0.12)
   plt.xscale("log")
   plt.xlabel(r"$p_{\mathrm{gossip}}$")
   plt.ylabel("Evidence lat. (min)")
   plt.title(fr"Typical regime: latency (mix={mix_level:g})")
   plt.grid(True, alpha=0.25, which="both")
   plt.legend(fontsize=6.5, frameon=True, ncol=3, loc="upper right")
   _savefig(outpath)
   plt.show()
def plot_latency_vs_overhead(df, mix_level, outpath, users=12000, det_thr=0.95):
   sub = df[(df["users"] == users) & np.isclose(df["mix_level"], mix_level)].copy()
# focus on where the protocol "works" reliably
   sub = sub[sub["detected_rate"].astype(float) >= det_thr].copy()
   sub = sub.sort_values(["adoption","p_gossip"])
   plt.figure(figsize=(3.35, 2.35))
   for a in sorted(sub["adoption"].unique()):
       s = sub[sub["adoption"] == a].copy()
       x = s["bpd_mean"].astype(float).to_numpy()
       y = s["lat_ev_med"].astype(float).to_numpy()
       pg = s["p_gossip"].astype(float).to_numpy()
       m = np.isfinite(x) & np.isfinite(y)
       plt.scatter(x[m], y[m], s=18, label=f"adopt={a:g}")
# label a few points to reduce clutter (only for smallest + largest pg)
       if m.any():
# pick low/hi pg indices
           idxs = np.argsort(pg[m])
           for j in [idxs[0], idxs[-1]]:
               xm, ym, pgm = x[m][j], y[m][j], pg[m][j]
               plt.text(xm, ym, f"{pgm:g}", fontsize=6)
   plt.xlabel("Overhead (B / msg)")
   plt.ylabel("Evidence lat. (min)")
   plt.title(fr"Typical regime (detâ‰¥{det_thr:g}): latency vs overhead (mix={mix_level:g})")
   plt.grid(True, alpha=0.25)
   plt.legend(fontsize=6.5, frameon=True, ncol=2, loc="upper right")
   _savefig(outpath)
   plt.show()
# ---- Generate a compact â€œ3-level mixingâ€ set of plots ----
for mix in [0.02, 0.03, 0.05]:
   plot_det_vs_pg(df_agg_typ_diag, mix, PLOTS_TYP / f"typ_det_mix{mix:g}")
   plot_lat_vs_pg(df_agg_typ_diag, mix, PLOTS_TYP / f"typ_lat_mix{mix:g}")
   plot_latency_vs_overhead(df_agg_typ_diag, mix, PLOTS_TYP / f"typ_lat_vs_over_mix{mix:g}", det_thr=0.95)
```

```python
# =========================
# Sweep C: Mixing Ã— p_gossip
# Goal: show that crosscut mixing controls detectability (latency/coverage),
# while overhead remains mostly governed by (adoption Ã— p_gossip).
# =========================
OUT_MIX = OUT_SWEEPS / "mixing_x_pgossip"
OUT_MIX.mkdir(parents=True, exist_ok=True)
BASE_MIX = dict(BASE_SMALL)
BASE_MIX.update(
   # window with fork in the middle
   hours=6.0,
   fork_hour=3.0,
   fork_duration=2.0,
   server_mode="permanent_fork",
   mode="sth_proof",
   # enough traffic for diffusion
   msg_rate_1to1=0.20,
   msg_rate_group=0.08,
   # keep adoption fixed so we isolate mixing+gossip effects
   adoption=0.3,
   # make fork harder than your current baseline if you want clearer separation
   fork_frac=0.05,
   # no censorship in this sweep
   p_drop_crosscut=0.0,
)
# We sweep the two mixing knobs together (same value)
MIX_LEVELS = [0.001, 0.005, 0.01]      # baseline/hard/very hard
PG_LEVELS  = [0.01, 0.05, 0.2]
USER_LEVELS = [2000, 12000]
rows = []
seed_base = 20000
for n in USER_LEVELS:
   for mix in MIX_LEVELS:
       sweep = dict(
           users=[n],
           p_contact_cross=[mix],
           p_group_cross=[mix],
           p_gossip=PG_LEVELS,
       )
       out_here = OUT_MIX / f"users{n}_mix{str(mix).replace('.','p')}"
       out_here.mkdir(parents=True, exist_ok=True)
       df_raw = run_grid(
           BASE_MIX, sweep,
           replicates=10,
           seed_base=seed_base,
           outroot=out_here,
           detection_policy="strict",
       )
       seed_base += 100
       rows.append(df_raw)
df_raw_mix = pd.concat(rows, ignore_index=True)
df_raw_mix.to_csv(OUT_MIX / "raw.csv", index=False)
group_cols = ["users","adoption","p_gossip","p_contact_cross","p_group_cross","fork_frac","server_mode","mode"]
df_agg_mix = aggregate(df_raw_mix, group_cols).sort_values(["users","p_contact_cross","p_gossip"])
df_agg_mix.to_csv(OUT_MIX / "agg.csv", index=False)
print("Wrote:", OUT_MIX)
df_agg_mix
```

```python
##compute IQR aggregation for bands
import numpy as np
import pandas as pd
from pathlib import Path
# ---------- group columns (must match your sweep parameters) ----------
group_cols_A = [
   "users","p_gossip","adoption",
   "server_mode","mode","fork_frac",
   "p_contact_cross","p_group_cross"
]
# ---------- NaN-safe aggregation: median/IQR conditioned on detection ----------
def aggregate_iqr_safe(df_raw: pd.DataFrame, group_cols):
   def _finite(x):
       x = np.asarray(x, dtype=float)
       return x[np.isfinite(x)]
   def q25(x):
       x = _finite(x)
       return float(np.quantile(x, 0.25)) if x.size else np.nan
   def q75(x):
       x = _finite(x)
       return float(np.quantile(x, 0.75)) if x.size else np.nan
   def med(x):
       x = _finite(x)
       return float(np.median(x)) if x.size else np.nan
   g = df_raw.groupby(group_cols, dropna=False)
   out = g.agg(
       reps=("run_dir", "count"),
       detected_rate=("detected", "mean"),
       # unconditional median (can be NaN if nobody detects)
       lat_ev_med_all=("lat_ev_min", "median"),
       # conditional stats computed safely (finite values only)
       lat_ev_med=("lat_ev_min", med),
       lat_ev_q25=("lat_ev_min", q25),
       lat_ev_q75=("lat_ev_min", q75),
       # overhead
       bpd_mean=("bytes_per_delivery", "mean"),
       bps_mean=("bytes_per_send", "mean"),
   ).reset_index()
   return out
# ---------- define df_agg_A ----------
df_agg_A = aggregate_iqr_safe(df_raw_A, group_cols_A).sort_values(["users","adoption","p_gossip"])
# Save for paper artifacts
OUT_HARD = Path(OUT_HARD)  # if OUT_HARD already exists; otherwise remove this line
df_raw_A.to_csv(OUT_HARD / "raw.csv", index=False)
df_agg_A.to_csv(OUT_HARD / "agg_iqr_safe.csv", index=False)
print("Wrote:", (OUT_HARD / "agg_iqr_safe.csv").resolve())
df_agg_A.head(10)
```

```python

# Paper-style "success-first" plots for Sweep A (hard)
# Focus: "What overhead buys fast + reliable detection?"
#  - Filters to points with detected_rate >= DET_THR
#  - Produces 3 plots:
#      (1) Context: detection vs p_gossip (log-x)
#      (2) Main: Pareto scatter (overhead bpd vs latency) with adoption-coded markers, annotated by p_gossip
#      (3) Requirement curve: min p_gossip to reach det>=DET_THR vs adoption
#
# EXPECTS df_agg_A to exist with columns:
#   users, adoption, p_gossip, detected_rate, lat_ev_med, bpd_mean
# Optional (for nicer reporting, not required):
#   lat_ev_q25, lat_ev_q75
# =========================
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
# ---- user knobs ----
U = 12000
DET_THR = 0.95
ADOPTIONS = (0.1, 0.3, 0.6)
# Where to write files (edit if you want)
PLOTS_SUCCESS = Path(OUT_HARD) / "plots_success"
PLOTS_SUCCESS.mkdir(parents=True, exist_ok=True)
def _as_float(s):
   return np.asarray(s, dtype=float)
def plot_A_success_bundle(df_agg_A, users=U, det_thr=DET_THR, adoptions=ADOPTIONS, outdir=PLOTS_SUCCESS):
   outdir = Path(outdir); outdir.mkdir(parents=True, exist_ok=True)
   # --- slice + normalize ---
   d = df_agg_A[df_agg_A["users"] == users].copy()
   if d.empty:
       raise RuntimeError(f"No rows for users={users}. Check df_agg_A['users'].unique().")
   for col in ["p_gossip","adoption","detected_rate","lat_ev_med","bpd_mean"]:
       if col not in d.columns:
           raise RuntimeError(f"Missing required column '{col}' in df_agg_A. Have: {list(d.columns)}")
   d["p_gossip"] = d["p_gossip"].astype(float)
   d["adoption"] = d["adoption"].astype(float)
   d["detected_rate"] = d["detected_rate"].astype(float)
   d["lat_ev_med"] = d["lat_ev_med"].astype(float)
   d["bpd_mean"] = d["bpd_mean"].astype(float)
   # =========================
   # (1) CONTEXT: Detection vs p_gossip (log-x)
   # =========================
   fig, ax = plt.subplots(1, 1, figsize=(3.3, 2.25))
   for a in adoptions:
       s = d[np.isclose(d["adoption"], a)].sort_values("p_gossip")
       if s.empty:
           continue
       x = _as_float(s["p_gossip"])
       y = _as_float(s["detected_rate"])
       m = np.isfinite(x) & np.isfinite(y)
       ax.plot(x[m], y[m], marker="o", linewidth=1.0, label=f"adopt={a:g}")
   ax.axhline(det_thr, linestyle="--", linewidth=1.0)
   ax.set_xscale("log")
   ax.set_ylim(-0.02, 1.02)
   ax.set_xlabel(r"$p_{\mathrm{gossip}}$")
   ax.set_ylabel("Detection prob.")
   ax.grid(True, alpha=0.25, which="both")
   ax.legend(fontsize=7, frameon=False)
   fig.tight_layout()
   fig.savefig(outdir / f"figA_context_detect_users{users}.pdf")
   fig.savefig(outdir / f"figA_context_detect_users{users}.png", dpi=260)
   plt.show()
   print("Wrote:", (outdir / f"figA_context_detect_users{users}.pdf").resolve())
   # =========================
   # Filter to SUCCESS points (detected_rate >= det_thr)
   # =========================
   good = d[d["detected_rate"] >= det_thr].copy()
   if good.empty:
       raise RuntimeError(f"No points satisfy detected_rate >= {det_thr}. Lower DET_THR or verify results.")
   # =========================
   # (2) MAIN: Pareto scatter (overhead vs latency), annotate p_gossip
   # =========================
   # adoption-coded markers (no manual colors)
   markers = {adoptions[i]: ["o","s","^","D","P","X"][i % 6] for i in range(len(adoptions))}
   fig, ax = plt.subplots(1, 1, figsize=(3.6, 2.4))
   # plot one adoption at a time so legend is clean
   for a in adoptions:
       s = good[np.isclose(good["adoption"], a)].copy()
       if s.empty:
           continue
       # Keep only finite pairs
       x = _as_float(s["bpd_mean"])     # overhead B/msg
       y = _as_float(s["lat_ev_med"])   # latency minutes
       m = np.isfinite(x) & np.isfinite(y)
       s = s.loc[m].copy()
       if s.empty:
           continue
       ax.scatter(
           _as_float(s["bpd_mean"]),
           _as_float(s["lat_ev_med"]),
           marker=markers[a],
           s=28,
           label=f"adopt={a:g}"
       )
       # annotate by p_gossip (compact)
       for _, row in s.iterrows():
           pg = float(row["p_gossip"])
           ax.annotate(
               f"{pg:g}",
               (float(row["bpd_mean"]), float(row["lat_ev_med"])),
               fontsize=6,
               xytext=(2, 2),
               textcoords="offset points"
           )
   # Short axis text for two-column
   ax.set_xlabel("Overhead (B / msg)")
   ax.set_ylabel("Latency (min)")
   ax.set_title(f"Success regime: detâ‰¥{det_thr:g} (N={users})", fontsize=9)
   ax.grid(True, alpha=0.25)
   ax.legend(fontsize=7, frameon=False, loc="best")
   fig.tight_layout()
   fig.savefig(outdir / f"figA_pareto_success_users{users}.pdf")
   fig.savefig(outdir / f"figA_pareto_success_users{users}.png", dpi=260)
   plt.show()
   print("Wrote:", (outdir / f"figA_pareto_success_users{users}.pdf").resolve())
   # =========================
   # (3) REQUIREMENT: min p_gossip to reach det>=thr, per adoption
   # =========================
   fig, ax = plt.subplots(1, 1, figsize=(3.3, 2.25))
   xs, ys = [], []
   for a in adoptions:
       s = d[np.isclose(d["adoption"], a)].sort_values("p_gossip")
       if s.empty:
           xs.append(a); ys.append(np.nan); continue
       s_ok = s[s["detected_rate"] >= det_thr]
       pg_min = float(s_ok["p_gossip"].min()) if not s_ok.empty else np.nan
       xs.append(a); ys.append(pg_min)
   xs = np.asarray(xs, float)
   ys = np.asarray(ys, float)
   m = np.isfinite(xs) & np.isfinite(ys)
   ax.plot(xs[m], ys[m], marker="o", linewidth=1.0)
   ax.set_yscale("log")
   ax.set_xlabel("Adoption")
   ax.set_ylabel(r"Min $p_{\mathrm{gossip}}$ for detâ‰¥" + f"{det_thr:g}")
   ax.grid(True, alpha=0.25, which="both")
   fig.tight_layout()
   fig.savefig(outdir / f"figA_requirement_users{users}.pdf")
   fig.savefig(outdir / f"figA_requirement_users{users}.png", dpi=260)
   plt.show()
   print("Wrote:", (outdir / f"figA_requirement_users{users}.pdf").resolve())
   print("\nAll success-first figures written to:", outdir.resolve())
# ---- run ----
plot_A_success_bundle(df_agg_A, users=U, det_thr=DET_THR, adoptions=ADOPTIONS, outdir=PLOTS_SUCCESS)
```

```python
# ===== Robust 2x2 compact figure: Sweep A + S1 + S2 + S3 (auto-detect / auto-load) =====

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from pathlib import Path

import os

# ---------- locate OUT_ROOT / OUT_SWEEPS robustly ----------

try:

    OUT_ROOT

except NameError:

    OUT_ROOT = Path.cwd()

OUT_ROOT = Path(OUT_ROOT)

OUT_SWEEPS = OUT_ROOT / "paper_sweeps"   # change if your notebook uses a different folder name

PLOTS_DIR = OUT_ROOT / "paper_plots"

PLOTS_DIR.mkdir(parents=True, exist_ok=True)

# ---------- helpers ----------

def first_existing_var(names):

    g = globals()

    for n in names:

        if n in g and isinstance(g[n], pd.DataFrame):

            return g[n], n

    return None, None

def try_load_agg(folder: Path):

    f = folder / "agg.csv"

    if f.exists():

        return pd.read_csv(f)

    return None

def pick_col(df, candidates):

    for c in candidates:

        if c in df.columns:

            return c

    return None

def safe_num(s):

    return pd.to_numeric(s, errors="coerce")

def ensure_mixing_index(df):

    if "mixing_index" in df.columns:

        return df

    if "p_contact_cross" in df.columns and "p_group_cross" in df.columns:

        df = df.copy()

        df["mixing_index"] = 0.5 * (safe_num(df["p_contact_cross"]) + safe_num(df["p_group_cross"]))

        return df

    return df

def plot_dual_iqr(ax, df, xcol, huecol, title,

                  xlog=False,

                  det_candidates=("det_prob","detected_rate","det_rate","detected"),

                  lat_med_candidates=("lat_med","lat_ev_med","lat_ev_min","latency_med"),

                  lat_q1_candidates=("lat_q1","lat_q1_min","lat_q1_med"),

                  lat_q3_candidates=("lat_q3","lat_q3_min","lat_q3_med"),

                  lat_iqr_candidates=("lat_iqr","lat_iqr_str","lat_iqr_tuple"),

                  det_ylim=(0,1.02)):

    if df is None or len(df) == 0:

        ax.set_title(title + " (no data)")

        ax.axis("off")

        return

    det_col = pick_col(df, det_candidates)

    lat_med_col = pick_col(df, lat_med_candidates)

    lat_q1 = pick_col(df, lat_q1_candidates)

    lat_q3 = pick_col(df, lat_q3_candidates)

    lat_iqr = pick_col(df, lat_iqr_candidates)

    ax2 = ax.twinx()

    if huecol not in df.columns or xcol not in df.columns:

        ax.set_title(title + " (missing columns)")

        ax.axis("off")

        return

    hues = sorted(df[huecol].dropna().unique(), key=lambda x: float(x) if str(x).replace(".","",1).isdigit() else str(x))

    for hv in hues:

        sub = df[df[huecol] == hv].copy()

        sub[xcol] = safe_num(sub[xcol])

        sub = sub.sort_values(xcol)

        xs = safe_num(sub[xcol]).to_numpy()

        dets = safe_num(sub[det_col]).to_numpy() if det_col else np.full_like(xs, np.nan, dtype=float)

        lmed = safe_num(sub[lat_med_col]).to_numpy() if lat_med_col else np.full_like(xs, np.nan, dtype=float)

        ax.plot(xs, dets, marker="o", linewidth=1.2, markersize=3,

                label=f"{huecol}={hv:g}" if isinstance(hv,(int,float,np.number)) else f"{huecol}={hv}")

        ax2.plot(xs, lmed, linestyle="--", linewidth=1.0, alpha=0.9)

        # IQR shading if available

        if lat_q1 and lat_q3:

            q1 = safe_num(sub[lat_q1]).to_numpy()

            q3 = safe_num(sub[lat_q3]).to_numpy()

            ax2.fill_between(xs, q1, q3, alpha=0.12, linewidth=0)

        elif lat_iqr:

            vals = sub[lat_iqr].astype(str).tolist()

            q1, q3 = [], []

            for s in vals:

                try:

                    s = s.strip().replace("[","").replace("]","").replace("(","").replace(")","")

                    a,b = s.split(",")

                    q1.append(float(a)); q3.append(float(b))

                except Exception:

                    q1.append(np.nan); q3.append(np.nan)

            q1 = np.array(q1); q3 = np.array(q3)

            if np.isfinite(q1).any() and np.isfinite(q3).any():

                ax2.fill_between(xs, q1, q3, alpha=0.12, linewidth=0)

    ax.set_title(title, pad=2)

    ax.set_ylim(*det_ylim)

    ax.set_ylabel("Det. prob.")

    ax.grid(True, which="both", linewidth=0.4, alpha=0.5)

    ax2.set_ylabel("Lat. (min)")

    ax2.grid(False)

    if xlog:

        ax.set_xscale("log", base=10)

    ax.legend(loc="upper left", frameon=False, handlelength=1.4, handletextpad=0.4,

              borderaxespad=0.2, labelspacing=0.2, fontsize=7)

# ---------- compact styling ----------

plt.rcParams.update({

    "font.size": 8,

    "axes.titlesize": 9,

    "axes.labelsize": 8,

    "xtick.labelsize": 7,

    "ytick.labelsize": 7,

    "legend.fontsize": 7,

})

# ---------- resolve dfs (in-memory first, then disk) ----------

# Sweep A

dfA, dfA_name = first_existing_var(["df_agg_A", "df_agg_hard", "df_agg_sweepA", "df_agg_sweepA_hard"])

if dfA is None:

    # common folder name you used: sweepA_hard_lowmix_dense (adjust if yours differs)

    for cand in ["sweepA_hard_lowmix_dense", "sweepA_hard", "sweepA"]:

        dfA = try_load_agg(OUT_SWEEPS / cand)

        if dfA is not None:

            dfA_name = f"loaded:{cand}"

            break

# S1 (targeted hardness)

dfS1, dfS1_name = first_existing_var(["df_agg_S1", "df_agg_targeted", "df_agg_victimfrac"])

if dfS1 is None:

    for cand in ["targeted_split_hardness", "sweepS1_targeted", "S1_targeted", "targeted"]:

        dfS1 = try_load_agg(OUT_SWEEPS / cand)

        if dfS1 is not None:

            dfS1_name = f"loaded:{cand}"

            break

# S2 (censorship)

dfS2, dfS2_name = first_existing_var(["df_agg_S2", "df_agg_censorship", "df_agg_drop"])

if dfS2 is None:

    for cand in ["censorship_robustness", "sweepS2_censorship", "S2_censorship", "censorship"]:

        dfS2 = try_load_agg(OUT_SWEEPS / cand)

        if dfS2 is not None:

            dfS2_name = f"loaded:{cand}"

            break

# S3 (mixing)

dfS3, dfS3_name = first_existing_var(["df_agg_S3", "df_agg_mix", "df_agg_mixing"])

if dfS3 is None:

    for cand in ["mixing_sensitivity", "sweepS3_mixing", "S3_mixing", "mixing"]:

        dfS3 = try_load_agg(OUT_SWEEPS / cand)

        if dfS3 is not None:

            dfS3_name = f"loaded:{cand}"

            break

print("Resolved dataframes:")

print("  A :", "OK" if dfA is not None else "MISSING", f"({dfA_name})")

print("  S1:", "OK" if dfS1 is not None else "MISSING", f"({dfS1_name})")

print("  S2:", "OK" if dfS2 is not None else "MISSING", f"({dfS2_name})")

print("  S3:", "OK" if dfS3 is not None else "MISSING", f"({dfS3_name})")

# Derive mixing_index if needed

if dfS3 is not None:

    dfS3 = ensure_mixing_index(dfS3)

# ---------- build figure ----------

fig = plt.figure(figsize=(6.8, 4.6), dpi=200, constrained_layout=True)

gs = fig.add_gridspec(2, 2, wspace=0.15, hspace=0.18)

axA  = fig.add_subplot(gs[0, 0])

axS1 = fig.add_subplot(gs[0, 1])

axS2 = fig.add_subplot(gs[1, 0])

axS3 = fig.add_subplot(gs[1, 1])

# Panel A: x=p_gossip, hue=adoption

if dfA is not None:

    plot_dual_iqr(axA, dfA, xcol="p_gossip", huecol="adoption",

                  title="A) Sweep A: $p_{gossip}$", xlog=True)

    axA.set_xlabel("$p_{gossip}$ (log)")

else:

    axA.set_title("A) Sweep A (missing)")

    axA.axis("off")

# Panel S1: x=fork_frac, hue=p_gossip (or adoption if needed)

if dfS1 is not None:

    hue = "p_gossip" if "p_gossip" in dfS1.columns else ("adoption" if "adoption" in dfS1.columns else None)

    x   = "fork_frac" if "fork_frac" in dfS1.columns else None

    plot_dual_iqr(axS1, dfS1, xcol=x, huecol=hue,

                  title="S1) Targeted hardness: victim fraction $f$", xlog=False)

    axS1.set_xlabel("Victim fraction $f$")

else:

    axS1.set_title("S1) Targeted hardness (missing)")

    axS1.axis("off")

# Panel S2: x=p_drop_crosscut, hue=p_gossip

if dfS2 is not None:

    hue = "p_gossip" if "p_gossip" in dfS2.columns else ("adoption" if "adoption" in dfS2.columns else None)

    x   = "p_drop_crosscut" if "p_drop_crosscut" in dfS2.columns else None

    plot_dual_iqr(axS2, dfS2, xcol=x, huecol=hue,

                  title="S2) Censorship: crosscut drop $p_{drop}$", xlog=False)

    axS2.set_xlabel("$p_{drop}$")

else:

    axS2.set_title("S2) Censorship (missing)")

    axS2.axis("off")

# Panel S3: x=mixing_index, hue=p_gossip

if dfS3 is not None:

    hue = "p_gossip" if "p_gossip" in dfS3.columns else ("adoption" if "adoption" in dfS3.columns else None)

    x   = "mixing_index" if "mixing_index" in dfS3.columns else None

    plot_dual_iqr(axS3, dfS3, xcol=x, huecol=hue,

                  title="S3) Mixing: crosscut index", xlog=True)

    axS3.set_xlabel("Mixing index (log)")

else:

    axS3.set_title("S3) Mixing (missing)")

    axS3.axis("off")

# Save

out_pdf = PLOTS_DIR / "fig_compact_4in1.pdf"

out_png = PLOTS_DIR / "fig_compact_4in1.png"

fig.savefig(out_pdf, bbox_inches="tight")

fig.savefig(out_png, bbox_inches="tight")

print("Wrote:", out_pdf)

print("Wrote:", out_png)

plt.show()
 
```

```python

def heatmap(df_agg, value_col, title, xcol="p_gossip", ycol="adoption"):
    hm = df_agg.copy()
    xs = sorted(hm[xcol].dropna().unique())
    ys = sorted(hm[ycol].dropna().unique())
    X = {x:i for i,x in enumerate(xs)}
    Y = {y:i for i,y in enumerate(ys)}
    Z = np.full((len(ys), len(xs)), np.nan)
    for _, r in hm.iterrows():
        Z[Y[r[ycol]], X[r[xcol]]] = r[value_col]
    import matplotlib.pyplot as plt
    plt.figure()
    plt.imshow(Z, aspect="auto", origin="lower")
    plt.xticks(range(len(xs)), [f"{x:g}" for x in xs], rotation=45)
    plt.yticks(range(len(ys)), [f"{y:g}" for y in ys])
    plt.xlabel(xcol); plt.ylabel(ycol); plt.title(title)
    plt.colorbar(); plt.tight_layout(); plt.show()

heatmap(df_agg_targ, "coverage_end_mean", "Targeted: coverage end")
heatmap(df_agg_targ, "t10_med", "Targeted: t10 (min)")
heatmap(df_agg_targ, "bpd_mean", "Targeted: bytes per delivery")
```
